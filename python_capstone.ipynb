```python
# Import required libraries for data manipulation and numerical operations
import pandas as pd
import numpy as np
```

```python
project_df = pd.read_csv('project.csv')
employee_df = pd.read_csv('employee.csv')
designation_df = pd.read_csv('designation.csv')
```

```python
project_df.head(), employee_df.head(), designation_df.head()
```

```python
merged_df = pd.merge(project_df, employee_df, on='ID', how='left')
merged_df
```

```python
# Task 4: Merge Project, Employee, and Designation dataframes into a single dataframe named Final
final_df = pd.merge(merged_df, designation_df, on='ID', how='left')
final_df
```

```python
final_df.to_csv('final_df.csv', index=False)
```

```python
final_df = pd.read_csv('final_df.csv')
final_df
```

```python
# Task 5: Add Bonus column giving 5% bonus for employees who finished projects
final_df['Bonus'] = np.where(
    final_df['Status'] == 'Finished',
    final_df['Cost'] * 0.05,
    0
)
```

```python
final_df[['ID', 'Project', 'Cost', 'Status', 'Bonus']].head()
```

```python
final_df.to_csv('final_df.csv', index=False)
```

```python
final_df = pd.read_csv('final_df.csv')
final_df[['ID', 'Status', 'Designation Level']].head()
```

```python
# Task 6:
# Reducing designation level by 1 for employees whose project status is Failed
final_df.loc[final_df['Status'] == 'Failed', 'Designation Level'] -= 1
```

```python
final_df[['ID', 'Status', 'Designation Level']].head(10)
```

```python
# Removing employees whose designation level is above 4 after demotion
final_df = final_df[
    ~((final_df['Status'] == 'Failed') & (final_df['Designation Level'] > 4))
]
```

```python
final_df.to_csv('final_df.csv', index=False)
```

```python
final_df = pd.read_csv('final_df.csv')
final_df[['First Name', 'Gender']].head()
```

```python
# Task 7:
# Adding title (Mr./Mrs.) to First Name based on Gender
# Dropping the Gender column after modification
final_df['First Name'] = np.where(
    final_df['Gender'] == 'M',
    'Mr. ' + final_df['First Name'],
    'Mrs. ' + final_df['First Name']
)
```

```python
final_df = pd.read_csv('final_df.csv')
final_df.head()
```

```python
total_cost_df = (
    final_df
    .groupby(['ID', 'First Name'])['Cost']
    .sum()
    .reset_index()
)
```

```python
total_cost_df
```

```python
total_cost_df.to_csv('total_project_cost.csv', index=False)
```

```python
final_df[['Age', 'Designation Level']].head()
```

```python
# Task 8:
# Promoting employees by increasing designation level by 1 if age is greater than 29
final_df['Designation Level'] = final_df.apply(
    lambda x: x['Designation Level'] + 1 if x['Age'] > 29 else x['Designation Level'],
    axis=1
)
```

```python
final_df[['Age', 'Designation Level']]
```

```python
# Task 9:
# Calculating total project cost for each employee
# Creating a new dataframe TotalProjCost with ID, First Name, and Total Cost
TotalProjCost = (
    final_df
    .groupby(['ID', 'First Name'])['Cost']
    .sum()
    .reset_index()
)

TotalProjCost.rename(columns={'Cost': 'Total Cost'}, inplace=True)
TotalProjCost
```
